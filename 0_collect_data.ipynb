````python
import json
import os

from github import Github
````

## Retrieve Issues on MOBs

````python
# Access token for GitHub
GITHUB_TOKEN = ""
g = Github(GITHUB_TOKEN)

repo = "pytorch/pytorch"
query = "is:issue linked:pr label:\"oncall: jit\" is:closed"

full_data = []

if os.path.exists('full_data.json'):
    with open('full_data.json', 'r') as f:
        full_data = json.load(f)
else:
    for issue in g.search_issues(query=f'repo:{repo} {query}'):
        full_data.append(issue.raw_data)
    with open('full_data.json', 'w') as f:
        json.dump(full_data, f)

print(len(full_data))
````

## Extract code from issue descriptions

````python
import re

pattern = re.compile(r'```.*?\r?\n(.*?)\r?\n```', re.DOTALL)

for raw_data in full_data:
    number = raw_data['number']
    body = raw_data['body']
    results = pattern.findall(body)
    for i, result in enumerate(results):
        print(number, i)
        with open(f'code/{number}_{i}.py', 'w') as f:
            f.write(result)
````
